Using your friends initial research, you'll delve into the Netflix data to see if you can determine whether movie lengths are actually getting shorter and explain some of the contributing factors, if any.

```python
# Netflix Movies Investigation
```

```python
## Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

## Load Data
netflix_data = pd.read_csv("netflix_data.csv")
print("Data loaded successfully.")

## Filter for Movies
# Correcting the DataFrame name from 'netflix_df' to 'netflix_data' for consistency
netflix_movies = netflix_data[netflix_data['type'] == 'Movie']
print("Movies filtered from the dataset.")

## Select Columns of Interest
# Correcting the DataFrame name to 'netflix_movies' and ensuring the column names are accurate
selected_columns = ['title', 'country', 'genre', 'release_year', 'duration']
netflix_movies = netflix_movies[selected_columns]
print("Relevant columns selected.")

## Identify Short Movies
# Ensure 'duration' is the correct column name and it contains the duration in minutes
short_movies = netflix_movies[netflix_movies['duration'] < 60]
print(f"Identified {len(short_movies)} short movies (duration < 60 mins).")

## Assign Colors Based on Genre
# Applying color assignment to 'netflix_movies' DataFrame
genre_to_color = {
    "Children": "orange",     
    "Documentaries": "green",  
    "Stand-Up": "purple"      
}
default_color = "grey"        
netflix_movies['color'] = netflix_movies['genre'].apply(lambda genre: genre_to_color.get(genre, default_color))
print("Colors assigned based on movie genres.")

## Display First 10 Color Assignments
print("First 10 color assignments:", netflix_movies['color'].head(10))

## Plotting the Data
# Creating a scatter plot for 'netflix_movies' DataFrame
plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=netflix_movies['color'])
plt.title("Netflix Movie Durations Over the Years")
plt.xlabel("Year of Release")
plt.ylabel("Duration (mins)")
plt.show()
print("Scatter plot created for movie durations.")

## Drawing Conclusions
conclusion = "Trend analysis required to determine duration changes."
print("Conclusion:", conclusion)
```
