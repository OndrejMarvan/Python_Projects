```python
import requests
import csv
from bs4 import BeautifulSoup
from IPython.display import display

# Send a request to the URL
url = 'https://www.techlib.cz/en/'
page = requests.get(url)

# Parse the HTML of the webpage
soup = BeautifulSoup(page.text, 'html.parser')

# Find the span element
element = soup.select_one('#content > div > div > div.col-xs-12.col-md-4.line-left > div.panel.panel-default.remove-bottom > div.panel-body.text-center.lead > span')

# Extract the text from the element
number = element.text

# Create a CSV string
csv_string = f'{number}\n'

# Display the CSV string as a file and allow downloading
display(csv_string, 'text/csv', 'numbers.csv')
```

```python
This code will send a request to the specified URL, retrieve the HTML of the webpage, parse it using BeautifulSoup, and find the span element using the select_one method. It will then extract the text from the element and create a CSV string containing the data.

The CSV string will be displayed as a file in the Jupyter Notebook, and you will be able to download the file by clicking on the download button that appears.

I hope this helps! Let me know if you have any other questions or if you need further assistance.
```

```python
import requests
import csv
import time
from bs4 import BeautifulSoup

# Open the CSV file in write mode
with open('numbers.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)

    # Repeat the request and retrieve the number every second
    while True:
        # Send a request to the URL
        url = 'https://www.techlib.cz/en/'
        page = requests.get(url)

        # Parse the HTML of the webpage
        soup = BeautifulSoup(page.text, 'html.parser')

        # Find the span element
        element = soup.select_one('#content > div > div > div.col-xs-12.col-md-4.line-left > div.panel.panel-default.remove-bottom > div.panel-body.text-center.lead > span')

        # Extract the text from the element
        number = element.text

        # Get the current time
        current_time = time.time()

        # Write the time and number to the CSV file
        writer.writerow([current_time, number])

        # Pause the loop for 1 second
        time.sleep(10)
```

```python
import requests
from bs4 import BeautifulSoup

# Send a request to the URL
url = 'https://www.globus.cz/cerny-most/cerpaci-stanice-a-myci-linka.html#!cookies-success'
page = requests.get(url)

# Parse the HTML of the webpage
soup = BeautifulSoup(page.text, 'html.parser')

# Find all the links on the page
links = soup.find_all('a')

# Open a file to write the data to
with open('links.txt', 'w') as file:
    # Write each link to the file
    for link in links:
        file.write(link.get('href') + '\n')
```
